```python
mov['rt_norms'] = np.round(mov['RottenTomatoes']/20,1)
mov['rt_user_norms'] = np.round(mov['RottenTomatoes_User']/20,1)
mov['meta_norms']=np.round(mov['Metacritic']/20,1)
mov['meta_user_norms']=np.round(mov['Metacritic_User']/2,1)
mov['IMDB_norms']=np.round(mov['IMDB']/2,1)
```

```python
mov.head().transpose()
```

```python
new_mov = mov[['STARS','RATING','rt_norms','rt_user_norms','meta_norms','meta_user_norms','IMDB_norms']]
```

```python
new_mov.head().transpose()
```

```python

fig,ax=plt.subplots(figsize=(12,8), dpi=200)
'''sns.kdeplot(mov['STARS'], shade=True, label='star')
sns.kdeplot(mov['RATING'], shade=True)
sns.kdeplot(mov['rt_norms'], shade=True)
sns.kdeplot(mov['rt_user_norms'], shade=True)
sns.kdeplot(mov['meta_norms'], shade=True)
sns.kdeplot(mov['meta_user_norms'], shade=True)
sns.kdeplot(mov['IMDB_norms'], shade=True)'''

def move_legend(ax, new_loc, **kws):
    old_legend = ax.legend_
    handles = old_legend.legendHandles
    labels = [t.get_text() for t in old_legend.get_texts()]
    title = old_legend.get_title().get_text()
    ax.legend(handles, labels, loc=new_loc, title=title, **kws)
    
    
sns.kdeplot(data=new_mov, clip=[0,5], shade=True, palette='Set1')
move_legend(ax, 'upper left')
```

```python
fig,ax = plt.subplots(figsize=(12,4), dpi=200)
sns.kdeplot(data=new_mov[['STARS','rt_norms']], clip=[0,5], shade=True, ax=ax, palette='Set1')
move_legend(ax,'upper left')
```
