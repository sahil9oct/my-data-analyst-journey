```python
def create_ages(mu=50, sigma=13, num_samples=100, seed=42):
      np.random.seed(seed)
      
      sample_ages = np.random.normal(loc=mu, scale=sigma,
                                     size=num_samples)
      sample_ages=np.round(sample_ages, decimals=0)
      
      return sample_ages
```
